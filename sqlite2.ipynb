```python
import sqlite3
```

```python
connection = sqlite3.connect("base_donne.db")
req = '''
  create table if not exists etudiant(
    etu_id integer primary key autoincrement,
    name text,
    prenom text,
    note real 
  )
'''
curseur = connection.cursor()
curseur.execute(req)
```

```python
requete = "insert into etudiant(name,prenom,note) values(?,?,?)"
nom = input("Entrer votre nom :")
prenom = input("Entrer votre prenom :")
note = float(input("Entrer votre note :"))
curseur.execute(requete, [nom,prenom,note])
connection.commit()
```
