```python
# créer une copie pour ne pas toucher au dataframe d'origine 
df = data.copy()
```

```python
# afficher la dataframe en entier
display(df) 
```

### Pratique 1: Suppression d'un attribut 

```python
# Procédez à la suppression de l'attribut "Nom"
df = df.drop(columns="Nom", axis=1)
df.head(2)
```

### Pratique 2: Technique de masquage 

```python
# Procéder au masquage des séries Siège et Ticket
# Essayez de repérer les informations qui peuvent conduire à une réidentification.
df["Siege"] = df["Siege"].map(lambda x: x[0:1]+"****")
# On pourrait décider de garder que les trois derniers chiffres pour la série Tickets
df.Ticket = df.Ticket.map(lambda x: "***" + x[-3:])
df
```

### Pratique 3: Suppression d'un enregistrement 

```python
# Procéder à la suppression des lignes contenant le prix le plus élevé ainsi que le prix le plus bas. 
prix_max = max(df["Prix"])
prix_min = min(df["Prix"])
condition = ((df["Prix"] != prix_max) & (df["Prix"] != prix_min))
df = df[condition]
df
```

```python
"""On voit que la colonne indexe a été touchée, on peut la réinitialiser avec la commande suivante"""
# reset index
df = df.reset_index(drop=True)

# Vérification: Afficher le dataframe
df
```

### Pratique 4: Tecnique de Géneralisation 

```python
def age_bins():
    fct = lambda x: "<18" if  x < 18 else "18-30" if x<31 else "31-50" if x<41 else "50>"
    return fct
```

```python
# Procéder à la généralisation de l’attribut "Age" et créer une nouvelle série catégorie age
df["Age_bins"] = df["Age"].map(age_bins())
df
```

```python
# supprimer la série âge
df.drop(columns = ["Age"], axis = 1, inplace = True)
df
```
